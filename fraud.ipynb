# 불필요한 컬럼 삭제

```python
drop_col = ['first', 'last', 'street', 'zip', 'trans_num', 'dob']

# 데이터프레임에서 불필요한 컬럼 제거
fraud_df = fraud_df.drop(columns=drop_col)
fraud_df.head()
```

# 2개의 피처 생성(위/경도사용, 통계적관점)

```python
# 사용자별 평균 및 표준편차를 계산해 거래 금액의 Z-score 생성
fraud_df['amt_zscore'] = fraud_df.groupby('cc_num')['amt'].transform(lambda x: (x - x.mean()) / x.std())
```

```python
fraud_df['cc_num_length'] = fraud_df['cc_num'].astype(str).apply(len)
# cc_num 카드번호 수가 달라서 이상치 처리를 할려고 했으나 6종류나 되어서 할수 없음
fraud_df['cc_num_length'].unique()
```

```python
fraud_df
```

```python
fraud_df = fraud_df.drop(columns='cc_num_length')
```

```python
fraud_df
```

```python
from geopy.distance import geodesic

# 카드 소유자와 상점 위치 간 거리 계산
fraud_df['distance'] = fraud_df.apply(lambda x: geodesic((x['lat'], x['long']), (x['merch_lat'], x['merch_long'])).kilometers, axis=1)

# 결과 확인
fraud_df[['lat', 'long', 'merch_lat', 'merch_long', 'distance']].head()
```

# 인코딩

```python
fraud_df.info()
```

```python
# 삭제할 컬럼 목록
drop_list2 = ['city', 'state', 'job', 'merchant', 'trans_date_trans_time']

# 데이터프레임에서 불필요한 컬럼 제거
fraud_df = fraud_df.drop(columns=drop_list2)
```

```python
fraud_df.info()
```

```python
# category와 gender 컬럼의 고유값 개수 출력
fraud_df[['category', 'gender']].nunique()
```

```python
# 원-핫 인코딩할 범주형 변수 지정
encode_col = ['category', 'gender']

# 원-핫 인코딩 수행
fraud_df = pd.get_dummies(fraud_df, columns=encode_col, drop_first=True)

# 결과 확인
fraud_df
```
