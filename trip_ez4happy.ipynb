```python
# datetime 형식으로 변환
data['tpep_pickup_datetime'] = pd.to_datetime(data['tpep_pickup_datetime'])
data['tpep_dropoff_datetime'] = pd.to_datetime(data['tpep_dropoff_datetime'])

# 주행 시간 계산 (분 단위)
data['trip_duration'] = (data['tpep_dropoff_datetime'] - data['tpep_pickup_datetime']).dt.total_seconds() / 60

# 데이터 확인
data
```

```python
# 상관 계수 계산
correlation = data[['trip_duration', 'trip_distance', 'fare_amount']].corr()
correlation
```

```python
# 'Credit Card'와 'Debit Card'를 모두 'Card'로 변경
data['payment_method'] = data['payment_method'].replace({'Credit Card': 'Card', 'Debit Card': 'Card'})
```
