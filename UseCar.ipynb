```python
import pandas as pd
import numpy as np
```

```python
cars_df = pd.read_csv('~/data/cars.csv')
brand_df = pd.read_csv('~/data/brand.csv')
```

```python
cars_df['brand'] = cars_df['title'].apply(lambda x: x.split()[0])
```

```python
cars_df
```

```python
cars_df['brand'] = cars_df['brand'].str.lower()
```

```python
missing_brands = cars_df[~cars_df['brand'].isin(brand_df['title'])]['brand'].unique()
```

```python
missing_brands
```

```python
merged_df = pd.merge(cars_df, brand_df, left_on='brand', right_on='title', how='left')
merged_df
```

```python
# 인코딩보다 결측치제거가 우선하는게 좋다고 생각함
merged_df.isna().mean()
```

```python
merged_df['Service history'] = merged_df['Service history'].fillna('Unknown')
```

```python
merged_df.isna().mean()
```

```python
threshold = 2  
merged_df = merged_df.dropna(thresh=merged_df.shape[1] - threshold)
```

```python
merged_df.isna().mean()
```

```python
#인코딩전에 컬럼수를 줄이기 위해
merged_df = merged_df.drop(columns=['title_y'])
```

```python
merged_df
```

```python
merged_df.describe()
```

```python
merged_df.info()
```

```python
merged_df[['title_x','Fuel type','Body type','Engine','Gearbox','Emission Class','Service history','brand','country']].nunique()
```

```python
# title_x는 brand 컬럼이 있으니 인코딩안하는게 좋을것 같음
merged_df = pd.get_dummies(merged_df,columns =['Fuel type','Body type','Engine','Gearbox','Emission Class','Service history','brand','country'],drop_first=True)
```

```python
merged_df.drop('title_x',axis = 1, inplace = True)
```

```python
merged_df
```

```python
#이상치가 없으므로 standarscaler사용함
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(merged_df)
scaled_df = pd.DataFrame(ss.transform(merged_df),columns = merged_df.columns)
```

```python
scaled_df
```

```python
# 결측치와 무한대값을 제거하기 위해
scaled_df.fillna(scaled_df.median(), inplace=True)
scaled_df.fillna(scaled_df.median(), inplace=True)
```

```python
from sklearn.decomposition import PCA
pca =  PCA(51)
pca_df = pd.DataFrame(pca.fit_transform(scaled_df))
pca_df
```

```python
(pca.explained_variance_ratio_).sum()
```

```python
#ss가 주성분이 너무 많아서 주성분분석을 rs로 다시함
from sklearn.preprocessing import RobustScaler
rs = RobustScaler()
rs.fit(merged_df)
scaled_r_df = pd.DataFrame(rs.transform(merged_df),columns = merged_df.columns)
scaled_r_df.fillna(scaled_r_df.median(), inplace=True)
scaled_r_df.fillna(scaled_r_df.median(), inplace=True)
```

```python
pca =  PCA(12)
pca_r_df = pd.DataFrame(pca.fit_transform(scaled_r_df))
(pca.explained_variance_ratio_).sum()
```

```python
#ss가 주성분이 너무 많아서  주성분분석을 mm로 다시함
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
mm.fit(merged_df)
scaled_m_df = pd.DataFrame(mm.transform(merged_df),columns = merged_df.columns)
scaled_m_df.fillna(scaled_m_df.median(), inplace=True)
scaled_m_df.fillna(scaled_m_df.median(), inplace=True)
```

```python
pca =  PCA(15)
pca_m_df = pd.DataFrame(pca.fit_transform(scaled_m_df))
(pca.explained_variance_ratio_).sum()
```

```python
# 가장 좋은 주성분분석은 rs사용한 주성분분석임, 주성분 12개로 기존 데이터의 정보 70% 이상을 가짐
pca_r_df
```
